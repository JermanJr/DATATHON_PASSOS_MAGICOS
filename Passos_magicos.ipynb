```python
def separa_ano_pedra(ano:object , df : pd.DataFrame):
    nome_colunas = list(df.filter(regex=ano))
    nome_colunas.insert(0,'NOME')
    df=df[nome_colunas].copy()
    return df
```

```python
def converter_lista_campos(lista_colunas : list, df :pd.DataFrame):
    for coluna in lista_colunas:
        df[coluna] = pd.to_numeric(df[coluna], errors='coerce')  # 'coerce' converte valores inválidos em NaN
```
